```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from pprint import pprint

import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from py_conway import Game  # nopep8
```

```python
my_game = Game(12, 12, random=True)

my_game.current_board
```

```python
my_game.start()

my_game.state
```

```python
plt.imshow(my_game.current_board, cmap='binary')
plt.show()
```

```python
my_game.run_generation()

plt.imshow(my_game.current_board, cmap='binary')
plt.show()
```

```python
board, ax = plt.subplots()
plt.title("Conway's Game of Life")

ylim, xlim = my_game.board_size

ax.set_xlim(0, xlim)
ax.set_ylim(0, ylim)

image = plt.imshow(my_game.current_board, animated=True)

def update(*args, **kwargs):
    my_game.run_generation()
    plt.title(f"Conway's Game of Life (Gen: {my_game.generations})")
    image.set_array(my_game.current_board)
    return image,

conway_animation = FuncAnimation(board, update, interval=50, blit=True)
plt.show()
```
